# Lab 8-1

```python
import datetime
today = datetime.date.today()
print(f'오늘의 날짜 : {today.year}년 {today.month}월 {today.day}일')
time = datetime.datetime.now()
if time.hour > 12:
    now_time = time.hour-12
    a = '오후'
else :
    now_time = time.hour
    a = '오전'
print(f'현재시간 : {a} {now_time}시 {time.minute}분 {time.second}초')
```

# Lab 8-3

## 1.

```python
import datetime as dt
thousand = dt.timedelta(days = 1000)
plus1000day = dt.datetime.now() + thousand
print(f'오늘로부터 1000일 뒤는 {plus1000day.year}년 {plus1000day.month}월 {plus1000day.day}일입니다')
```

## 2.

```python
years, months, days = map(int, input('처음으로 사귄 연도와 얼, 일을 입력하시오: ').split())
hundred = dt.timedelta(days=100)
first = dt.datetime.now()
first.replace(year = years, month = months, day = days)
plus100day = first + hundred
print(f'100일 기념일은 : {plus100day.year}년 {plus100day.month}월 {plus100day.day}일입니다.')
```

## 1.

```python
import random as rd

lotto_list = list(range(1, 46))
print('이번 주의 추천 로또 번호')
for i in range(5):
    rd.shuffle(lotto_list)
    lotto = lotto_list[:6]
    lotto.sort()
    print(lotto)
```

## 2.

```python
lotto_list = list(range(1, 46))
rd.shuffle(lotto_list)
lotto = lotto_list[:6]
bonus = lotto_list[6:7]
print('이번 주의 당첨 로또 번호 + 보너스')
print(lotto, '+', bonus)
```
