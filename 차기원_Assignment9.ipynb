```python
import datetime
today = datetime.date.today()
print(f'오늘의 날짜 : {today.year}년 {today.month}월 {today.day}일')
time = datetime.datetime.now()
if time.hour > 12:
    now_time = time.hour-12
    a = '오후'
else :
    now_time = time.hour
    a = '오전'
print(f'현재시간 : {a} {now_time}시 {time.minute}분 {time.second}초')
```

## 1.

```python
import datetime as dt
thousand = dt.timedelta(days = 1000)
plus1000day = dt.datetime.now() + thousand
print(f'오늘로부터 1000일 뒤는 {plus1000day.year}년 {plus1000day.month}월 {plus1000day.day}일입니다')
```

## 2.

```python
years, months, days = map(int, input('처음으로 사귄 연도와 얼, 일을 입력하시오: ').split())
hundred = dt.timedelta(days=100)
first = dt.datetime.now()
first.replace(year = years, month = months, day = days)
plus100day = first + hundred
print(f'100일 기념일은 : {plus100day.year}년 {plus100day.month}월 {plus100day.day}일입니다.')
```

```python
# 발생 예외: PermissionError
# 설명: 운영체제 수준에서 파일이 '읽기 전용'으로 설정되어 있어, 파이썬에서 쓰기 모드('w' 또는 'a')로 파일을 열려고 시도할 때 권한이 거부되며 발생하는 예외입니다.

import os
import stat

file_name = "my_hello.txt"


with open(file_name, 'w') as f:
    f.write("Test data")

os.chmod(file_name, stat.S_IREAD)

try:
    with open(file_name, 'w') as f:
        f.write("Hello Python")
except Exception as e:
    print(f"발생한 예외: {type(e).__name__}")
    print(f"예외 메시지: {e}")
```

```python
file_name = "my_hello.txt"

try:
    with open(file_name, 'w') as f:
        f.write("Hello Python")
    print("파일이 성공적으로 저장되었습니다.")
    
except PermissionError:
    print("오류: 파일이 '읽기 전용' 상태이거나 쓰기 권한이 없습니다. 속성을 확인해주세요.")
```

```python
f = open('hello.txt', 'w')
f.write("Hello.\nHi, there.\n")
f.close()

print("--- (1) 실행결과 ---")
print("hello.txt 파일 :")

f = open('hello.txt', 'r')
content = f.read()
print(content, end='')
f.close()

print("\n")

print("--- (2) 실행결과 ---")

f = open('hello.txt', 'a')
f.write("Welcome to Python!.\n")
f.close()

print("hello.txt 파일 :")

f = open('hello.txt', 'r')
new_content = f.read()
print(new_content, end='')
f.close()
```
